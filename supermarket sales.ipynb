# Q7 Get Overall Statistics About the DataFrame

```python
df.describe()
```

```python
# Gross Income column is in right skewed and right side of the tail this is because mean>median.
```

```python
gross_income_mean=df['gross income'].mean()
gross_income_median=df['gross income'].median()
```

```python
plt.figure(figsize=(5,5))
plt.hist(df['gross income'],bins=10)
plt.axvline(gross_income_mean,color='green')
plt.axvline(gross_income_median,color='red')
plt.title('Gross Income Frequency distribution')
plt.xlabel('Gross Income')
plt.ylabel('Frequency')
plt.show()
```
